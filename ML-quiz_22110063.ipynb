Q1. After performing matrix multiplication on two matrices of dimensions (n,k) and (k,m), the resultant matrix (M) is of the dimension (m,n). 
For each element in M, we are performing k multiplications. Since there are n * m elements in M, the total number of multiplications required is n * k * m.
For each element in M, we need to sum up k products, so we are doing k-1 additions for each element. Since there are n * m elements, the total number of additions we need to perform is n * m * (k-1).

```python
# Q2.

import numpy as np
import time

def matrix_multiply(mat1, mat2):
    rows_mat1 = len(mat1)
    cols_mat1 = len(mat1[0])
    rows_mat2 = len(mat2)
    cols_mat2 = len(mat2[0])

    if cols_mat1 != rows_mat2:
        raise ValueError("Cannot multiply")

    result = [[0 for _ in range(cols_mat2)] for _ in range(rows_mat1)]

    for i in range(rows_mat1):
        for j in range(cols_mat2):
            for k in range(cols_mat1):
                result[i][j] += mat1[i][k] * mat2[k][j]

    return result

n, k, m = 100, 100, 100

matrix_A = np.random.rand(n, k)
matrix_B = np.random.rand(k, m)

list_A = matrix_A.tolist()
list_B = matrix_B.tolist()

start_time = time.time()
result_lists = matrix_multiply(list_A, list_B)
lists_time = time.time() - start_time

start_time = time.time()
result_numpy = np.dot(matrix_A, matrix_B)
numpy_time = time.time() - start_time

print(f"Time taken by lists of lists approach: {lists_time:.6f} seconds")
print(f"Time taken by NumPy approach: {numpy_time:.6f} seconds")
```

```python
# Q9.

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.5, 100.0, 0.5)

y1 = x
y2 = x**2
y3 = x**3 / 100
y4 = np.sin(x)
y5 = np.sin(x) / x
y6 = np.log(x)
y7 = np.exp(x)

plt.figure(figsize=(12, 8))

plt.subplot(4, 2, 1)
plt.plot(x, y1)
plt.title('y = x')

plt.subplot(4, 2, 2)
plt.plot(x, y2)
plt.title('y = x^2')

plt.subplot(4, 2, 3)
plt.plot(x, y3)
plt.title('y = x^3 / 100')

plt.subplot(4, 2, 4)
plt.plot(x, y4)
plt.title('y = sin(x)')

plt.subplot(4, 2, 5)
plt.plot(x, y5)
plt.title('y = sin(x) / x')

plt.subplot(4, 2, 6)
plt.plot(x, y6)
plt.title('y = log(x)')

plt.subplot(4, 2, 7)
plt.plot(x, y7)
plt.title('y = e^x')

plt.tight_layout()
plt.show()
```

```python
#13Q. 

import numpy as np

def custom_argmin(arr):
    min_index = 0 
    for i in range(1, len(arr)):
        if arr[i] < arr[min_index]:
            min_index = i

    return min_index

array = np.array([5, 3, 8, 1, 7])
index_min = custom_argmin(array)
print("Index of minimum element (Custom function):", index_min)
index_np_argmin = np.argmin(array)
print("Index of minimum element (np.argmin):", index_np_argmin)
```

```python
#6Q. 

import sympy as sp

x, y = sp.symbols('x y')
f = x**2 * y + y**3 * sp.sin(x)
partial_x = sp.diff(f, x)
partial_y = sp.diff(f, y)
analytical_gradient = (partial_x, partial_y)
print("Analytical Gradient:", analytical_gradient)
```
